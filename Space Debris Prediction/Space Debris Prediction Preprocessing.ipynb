```python
import datetime

def calc_epoch(epoch):
    # epoch year
    if (0 <= int(epoch[:2]) <= 20):
        year = int('20' + epoch[:2])
    else:
        year = int('19' + epoch[:2])
    
    epoch = epoch[2:].split('.')
    # day of year
    day_of_year = int(epoch[0]) - 1
    date = datetime.date(year, 1, 1) + datetime.timedelta(day_of_year)
    
    # extract month and day
    month = int(date.month)
    day = int(date.day)
    
    # decimal hour
    hour = int(float('0.' + epoch[1]) * 24.0)
    # decimal minute
    minute = int((float('0.' + epoch[1]) * 24.0 - hour) * 60.0)
    # decimal second
    sec = ((float('0.' + epoch[1]) * 24.0 - hour) * 60.0 - minute) * 60.0
    
    return year, month, day, hour, minute, sec

calc_epoch("97223.23211")
```

```python
def tle2coes(tle_filename):
    with open(tle_filename, 'r') as f:
        lines = f.readlines()
    line1 = lines[0].strip().split()
    line2 = lines[1].strip().split()
    
    # epoch (year and day)
    epoch = line1[3]
    year, month, day = calc_epoch(epoch)
    
    # inclination
    i = float(line2[2])
    # right ascensison of the ascending node
    raan = float(line2[3])
    # eccentricity
    e_string = line2[4]
    e = float('0.' + e_string)
    # argument of perigee
    aop = float(line2[5])
    # mean anomaly
    M = float(line2[6])
    # mean motion
    n = float(line2[7][:11]) # revs/day
    
    return i, raan, e, aop, M, n, [year, month, day]
```
